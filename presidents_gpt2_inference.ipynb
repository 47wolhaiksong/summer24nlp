```python
from transformers import AutoTokenizer, AutoModelForCausalLM, Trainer, TrainingArguments
from datasets import Dataset
import torch
import pandas as pd
```

```python
checkpoint = 'adarksky/president-gpt2'
```

```python
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
tokenizer.pad_token = tokenizer.eos_token
```

```python
# Load model for causal language modeling
model = AutoModelForCausalLM.from_pretrained(checkpoint)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python
input_text = "We are all"
input_ids = tokenizer.encode(input_text, return_tensors='pt').to(device)
```

```python
# Generate encoded text
with torch.no_grad():
  outputs = model.generate(input_ids, min_length=50, max_length=100, num_return_sequences=1, do_sample=True, temperature=1.9, top_k=50)

# Decode and print the generated text
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
```
